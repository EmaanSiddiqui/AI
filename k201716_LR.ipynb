```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

PART A

```python
dicti= {'TV' : [230.1, 44.5, 17.2, 151.5, 180.8], 
        'Radio' : [37.8, 39.3, 45.9, 41.3, 10.8],
        'Newspaper' : [69.2, 45.1, 69.3, 58.5, 58.4],
        'Sales' : [22.1, 10.4, 9.3, 18.5, 12.9]}
        
df= pd.DataFrame(dicti)
df
```

```python
X = df.iloc[:,0:3]
Y = df.iloc[:,3]
```

```python
X
```

```python
Y
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X,Y,test_size = 0.2)
model = LinearRegression()
model.fit(xtrain,ytrain)
pred = model.predict(xtest)
print("Actual output:", ytest.to_numpy()[0] ,"\nPredicted output: ",pred[0])
```

```python
print("Increase in single sales will be resulted from increase of TV,radio and newspaper in respected values ",model.coef_)
print("y intercept: ", model.intercept_)
```
